```python
import requests
r = requests.get("https://gamma-api.polymarket.com/events?closed=true&end_date_min=2024-12-10T00:00:00Z&end_date_max=2024-12-14T00:00:00Z&limit=500")
```

```python
response = r.json()
len(response)
```

```python
response
```

```python
TGA_event = {}
for event in response:
    if "Game of the Year" in event['title']:
        print(event)
        TGA_event[event['id']] = event

len(TGA_event)
```

```python
import json
with open("TGA_event.txt", "w", encoding="utf-8") as f:
    f.write(json.dumps(TGA_event, indent=4, ensure_ascii=False))

print("finish")
```
