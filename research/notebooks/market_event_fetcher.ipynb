```python
import requests
import json

# Initialize parameters for pagination
params = {
    "closed": "false",
    "limit": 100,  # API's maximum supported limit per page
    "offset": 0    # Start from the first item
}

all_events = []  # To store all fetched events

# Loop until there are no more events returned in the response
while True:
    # Make the GET request with parameters
    r = requests.get("https://gamma-api.polymarket.com/events", params=params)
    response = r.json()
    
    # Break the loop if no more events are returned
    if not response:
        break

    # Add events from this page to the main list
    all_events.extend(response)
    
    # Increase the offset to get the next set of results
    params["offset"] += params["limit"]

# Now, filter and print the relevant information
for event in all_events:
    #if 'NBA' in event['title']:
    if 'Will Bitcoin hit $90k or $100k first?' in event['title']:
        print(f"Event ID: {event['id']}")
        print(f"Title: {event['title']}")

        # Print market ids, questions, outcome prices, and clobTokenIds for active markets
        for market in event['markets']:
            if not market.get("closed"):  # Check if the market is active
                print(f"Market ID: {market['id']}")
                print(f"Question: {market['question']}")
                
                if 'outcomePrices' in market:
                    print("Outcome Prices:", market['outcomePrices'])
                
                if 'conditionId' in market:
                    print("conditionId:", market['conditionId'])

                if 'slug' in market:
                    print("slug:", market['slug'])
                    
                if 'clobTokenIds' in market:
                    print("Clob Token IDs:", market['clobTokenIds'])
                
                print("=====")
```
